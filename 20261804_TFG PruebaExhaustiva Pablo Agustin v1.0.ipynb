```python
import itertools
from collections import deque, defaultdict

# CASO PARA DISTINTO DEPÓSITO DE SALIDA Y ENTRADA

origenes = ["Pallet Town", "Celadon City", "Fuschia City"]
destino = "Vermillion City"

def mapear(E):
    aristas = defaultdict(list)
    for arista_aux in E:
        u, v = tuple(arista_aux)
        aristas[u].append(v)
        aristas[v].append(u)
    
    return aristas

def bfs(E, inicio, objetivo):

    map = E

    cola = deque([inicio])
    visitados = {inicio}
    padre = {inicio: None}

    while cola:
        u = cola.popleft()

        if u == objetivo:
            break

        for v in map[u]:
            if v not in visitados:
                visitados.add(v)
                padre[v] = u
                cola.append(v)

    if objetivo not in padre:
        return None  # no hay camino

    camino = []
    actual = objetivo
    while actual is not None:
        camino.append(actual)
        actual = padre[actual]

    camino.reverse()
    return camino

def resolver(E, origenes, destino):
    map = mapear(E)

    flag = 1
    mejor_coste = 0
    mejores = []

    for inicio, fin in itertools.permutations(origenes, 2):
        c1 = bfs(map, inicio, destino)
        c2 = bfs(map, destino, fin)

        if c1 is None or c2 is None:
            continue

        ruta_total = c1 + c2[1:]
        coste = len(ruta_total) - 1

        if coste < mejor_coste or flag == 1:
            mejor_coste = coste
            mejores = [(inicio, fin, ruta_total, coste)]
            flag = 0
        elif coste == mejor_coste:
            mejores.append((inicio, fin, ruta_total, coste))

    return mejor_coste, mejores

coste, soluciones = resolver(E, origenes, destino)

print(f"Coste óptimo: {coste}\n")

for i, (inicio, fin, ruta, coste_sol) in enumerate(soluciones, start=1):
    print(f"Solución {i}")
    print(f"  Inicio : {inicio}")
    print(f"  Destino: {destino}")
    print(f"  Final  : {fin}")
    print(f"  Ruta   : {' -> '.join(ruta)}")
    print(f"  Coste  : {coste_sol}")
```
